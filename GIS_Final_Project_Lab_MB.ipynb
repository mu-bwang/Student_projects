````python
#Homework 1 Solution

import laspy
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the LiDAR file
file_path = "/content/drive/MyDrive/Colab Notebooks"
las_file = laspy.read('1690_1125.las')

# Get the x, y, and z coordinates of the LiDAR points
x = las_file.x
y = las_file.y
z = las_file.z

# Create a 3D scatter plot of the LiDAR point cloud
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, s=0.1, c=z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
````
